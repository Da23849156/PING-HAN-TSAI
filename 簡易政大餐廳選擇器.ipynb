```python
from datetime import * #匯入datatime模組所有東西
import random #匯入random模組

ss=0 #建立變項
b=1 #建立變項
c=1 #建立變項
td=date.today() #將今天的日期指派為td
dict1={'台式料理':['左撇子','敏忠小吃'],'義式料理':['里克德義廚房','Schumann'],'日式料理':['金鮨日式料理','加賀日本料理']} #建立料理-餐廳字典(中價位)
dict1a={'台式料理':['楊記小吃','悅來'],'義式料理':['提洛斯','Pasta2go'],'日式料理':['鬼匠','東京小城']}#建立料理-餐廳字典(低價位)

#建立餐廳-推薦菜單字典
dict2={'左撇子':['梅子雞丁飯','糖醋雞丁飯'],'敏忠小吃':['紅燒魚','糖醋排骨'],'加賀日本料理':['綜合生魚丼','雞腿丼'],'楊記小吃':['香滷大腸飯','麻辣臭豆腐鴨血'],'悅來':['牛肉麵','餛飩肉臊乾麵'],'提洛斯':['白酒蛤蠣義大利麵','青醬雞柳義大利麵'],'Pasta2go':['大蒜蝦子橄欖油麵','松子青醬麵'],'鬼匠':['豚骨拉麵','味噌拉麵'],'里克德義廚房':['干貝燻鮭義大利麵','黑松露野菇魷魚香料義大利麵'],'Schumann':['清炒時蔬義大利麵','蒜辣香腸義大利麵'],'金鮨日式料理':['政男蓋飯','政妹蓋飯'],'東京小城':['叉燒拉麵','豬排丼']}
#建立餐廳-平均價位字典
dict3={'左撇子':'200','敏忠小吃':'200','加賀日本料理':'200','楊記小吃':'100','悅來':'100','提洛斯':'100','Pasta2go':'100','鬼匠':'100','里克德義廚房':'200','Schumann':'200','金鮨日式料理':'200','東京小城':'200'}

#建立料理的list
list1=['台式料理','義式料理','日式料理']

#印出今天的日期以及今天是星期幾，以及開場文
print('今天是'+str(td.year)+'年'+str(td.month)+'月'+str(td.day)+'號星期'+str(td.isoweekday())+'，您對選擇要吃什麼有問擾嗎？讓本系統來幫助你吧！')


while True: #建立迴圈
        money=input('請輸入本次用餐的預算(只接受整數)，若您想退出本程式請輸入esc:') #請受試者輸入預算並將輸入回應指派為money
        if money.isdecimal(): #判斷money是否只有數字
                money=int(money) #將money轉為整數形式並且指派成money
                print('您本次的預算為',money,'元') #印出money的結果
                if 30<money<=150: #判斷money是否介於30-150
                        eat=dict1a #如果是，將dict1a字典指派成eat
                        
                elif money>150: #判斷money是否大於150
                        eat=dict1 #如果是，將dict1字典指派成eat
                else: #如果上述判斷都不符合
                        print('不好意思您的預算可能沒辦法吃到本系統裡面的餐廳，本系統將自動關閉') #印出文字
                        b=3 #將3指派到b
                        break #跳出迴圈
                
                print('\n以下為政大所擁有的餐廳類型，可供您挑選：\n') #印出文字
                while ss<3: #建立迴圈，判斷為ss<3
                        print('%s'%(list1[ss])) #印出index為ss的list1的內容
                        print() #印出一行
                        ss+=1   #將ss+1指派為ss
                break #跳出迴圈

        
        elif money=='esc': #判斷是否money是esc
                b=3 #將3指派為b
                break #跳出迴圈
        
        else: #如果money不是數字或是esc
                print('請輸入整數喔') #印出文字
                       

while b==1: #建立回圈，判斷b是否1
        a=input('您是否要隨機挑選料理類型？隨機請填1，輸入esc退出本程式:') #請受試者輸入，將輸入結果指派為a
        if a=='1': #判斷a是否為1
                while c==1: #建立迴圈，判斷c是否為1
                        name=random.choice(list1) #隨機從list1取出一個內容並指派為name
                        print('\n'+'-'*50) #印出50個-
                        print('恭喜您抽到'+name) #印出抽到什麼name
                        print('-'*50) #印出50個-
                        while True: #建立迴圈
                                z=input('\n您是否滿意系統挑選的類型？是請填1，否請填2(系統會重新挑選):') #請受試者輸入是否滿意，並將結果指派為z
                                if z=='1': #判斷z是否為1
                                        print('\n系統為您挑選的類型是'+name+'\n') #如果是，印出挑選的name
                                        c=2 #將2指派為c
                                        b=2 #將2指派為b
                                        break #跳出迴圈
                                elif z=='2': #判斷z是否為2
                                        print('系統將為您重新挑選\n')#如果是，印出文字
                                        list1.remove(name) #將原本抽到的name資料從list1去除
                                        if len(list1)==0: #判斷list1長度是否為0
                                                print('已經挑過一輪了啦！趕快決定!')#如果是，印出文字
                                                list1=['台式料理','義式料理','日式料理'] #重新建立list1變項
                                                break #跳出迴圈
                                        else:    #list1長度不等於0
                                                break #跳出迴圈
                                else: #如果z不等於1或2
                                        print('請按照指定格式輸入，謝謝您\n') #印出文字
                                        
       
        elif a=='esc': #判斷a是否是esc
                b=3 #如果是，將b=3
        else: #如果a不符合任何判斷
                 print('請按照指定格式輸入，謝謝您\n') #印出文字

if b==2: #判斷是否b=2
        print('在政大周邊有賣「'+name+'」的餐廳有',end=' ') #如果是印出文字
        h=eat.get(name) #將eat字典中將name的value抓出並指派為h
        for num in range(0,2): #建立迴圈，跑2次
                print(h[num],end=' ') #印出index為num的h的內容
                
        while True: #建立迴圈
                sq=input('請輸入您想要的餐廳名？(隨機挑選請輸入helpp)：') #請受試者輸入想要的餐廳名並指派為sq
                if sq=='helpp': #判斷sq是否為helpp
                        q=int(random.randint(0,1)) #隨機製造0-1的數字並指派為q
                        print('\n系統幫您選擇的是',h[q]) #印出index為q的h的值
                        print('-'*50) #印出50個-
                        print(h[q],'的平均價位為',dict3.get(h[q]),'\n') #印出index為q的h的值以及index為q的dict3的value
                        print(h[q],'的推薦菜單有',dict2.get(h[q]),'\n')#印出index為q的h的值以及index為q的dict2的value
                        print('-'*50) #印出50個-
                        break #跳出迴圈
                elif (sq!=h[0]) and (sq!=h[1]): #判斷是否sq不等於index是1且0的h的值
                        print('\n請按照指定格式輸入，謝謝您\n') #印出文字
                        
                elif h[0]!=sq or h[1]==sq: #判斷是否sq不等於index為0的h的值或等於 index為1的h的值
                        print('\n您選擇的是'+sq) #印出文字
                        print('-'*50) #印出-
                        print('\n'+sq+'的平均價位為',dict3.get(h[1]),'\n') #印出sq和index為1的dict3的value
                        print('\n'+sq+'的推薦餐單有',dict2.get(h[1]),'\n') #印出sq和index為1的dict2的value
                        print('-'*50) #印出-
                        break #跳出迴圈
 
                else: #如果不符合上述判斷
                        print('\n您選擇的是'+sq) #印出sq
                        print('\n'+'-'*50) #印出-
                        print('\n'+sq+'的平均價位為',dict3.get(h[0]),'\n') #印出sq和index為0的dict3的value
                        print('\n'+sq+'的推薦餐單有',dict2.get(h[0]),'\n') #印出sq和index為0的dict2的value
                        print('\n'+'-'*50) #印出-
                        break #跳出迴圈
elif b==3: #判斷b是否=3
        print('感謝您的支持') #如果是，印出文字
        
else: #如果b不符合上述判斷
        print('輸入錯誤請重新執行程式') #印出文字

        
                

```
